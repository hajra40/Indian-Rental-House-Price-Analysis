```python
#Merge all 3
merged_data = pd.concat([delhi, mumbai, pune], axis=0)
```

```python
merged_data.drop(['priceSqFt'],axis=1, inplace=True)
```

```python
merged_data["numBalconies"] = merged_data["numBalconies"].replace(["null", None], 0).fillna(0)
merged_data["isNegotiable"] = merged_data["isNegotiable"].replace(["Negotiable"], 1).fillna(0)

merged_data['isNegotiable'].unique()
```

```python
merged_data["numBathrooms"].unique()
```

```python
median_baths = merged_data["numBathrooms"].median()
merged_data["numBathrooms"] = merged_data["numBathrooms"].fillna(median_baths)
merged_data["numBathrooms"].unique()
```

```python
import re

# make sure column is str
s = merged_data['house_size'].astype(str)

# 1) Normalize: replace common unicode NBSP and remove commas, unify text
s_clean = (
    s
    .str.replace('\u00A0', ' ', regex=False)   # NBSP -> space
    .str.replace(',', '', regex=False)         # remove commas in numbers
    .str.lower()                               # unify case
    .str.strip()
)

# store cleaned string (optional, for inspection)
merged_data['house_size_clean'] = s_clean

# 2) Primary extraction: only capture numbers that are followed by 'sq ft' (robust to 'sq ft', 'sqft', 'sq. ft' etc.)
merged_data['house_size_num'] = s_clean.str.extract(r'([0-9]+(?:\.[0-9]+)?)\s*(?:sq(?:\.|\s)?ft|square\s?feet|sqft)', flags=re.IGNORECASE)[0]

# 3) Fallback #1: if still missing, try any large number (3+ digits) in the string (likely sqft)
mask_missing = merged_data['house_size_num'].isna()
merged_data.loc[mask_missing, 'house_size_num'] = s_clean[mask_missing].str.extract(r'([0-9]{3,}(?:\.[0-9]+)?)')[0]

# 4) Fallback #2 (optional): if still missing but the value is like '400' (2-3 digits), pick first number
mask_missing = merged_data['house_size_num'].isna()
merged_data.loc[mask_missing, 'house_size_num'] = s_clean[mask_missing].str.extract(r'([0-9]+(?:\.[0-9]+)?)')[0]

# 5) Convert to numeric
merged_data['house_size_num'] = pd.to_numeric(merged_data['house_size_num'], errors='coerce')

# 6) Compute rentPerSqft correctly (price divided by area)
merged_data['rentPerSqft'] = np.where(
    (merged_data['house_size_num'].notna()) & (merged_data['house_size_num'] > 0),
    merged_data['price'] / merged_data['house_size_num'],
    np.nan
)

# 7) Diagnostics: show rows still problematic (very small area or NaN)
problem_rows = merged_data[
    (merged_data['house_size_num'].isna()) |
    (merged_data['house_size_num'] < 50) |   # suspiciously small areas
    (merged_data['rentPerSqft'] > 10000)     # suspiciously high rentPerSqft
][['house_size', 'house_size_clean', 'house_size_num', 'price', 'rentPerSqft']]

print("Number of problematic rows:", len(problem_rows))

if len(problem_rows) == 0:
    display(merged_data[["house_size_num","rentPerSqft", "house_size"]])
else: 
    display(problem_rows.head(50))   # if in notebook; else print(problem_rows.head(50).to_string())
```

```python
merged_data["house_type_clean"] = merged_data["house_type"].str.strip().str.title()
merged_data["numBHK"] = merged_data["house_type_clean"].str.extract(r"(\d+)\s*Bhk", flags=re.IGNORECASE)
merged_data["numBHK"] = merged_data["numBHK"].fillna(0).astype(int)
def extract_property_type(x):
    x = x.lower()
    if "apartment" in x:
        return "Apartment"
    elif "independent floor" in x:
        return "Independent Floor"
    elif "independent house" in x:
        return "Independent House"
    elif "villa" in x:
        return "Villa"
    elif "penthouse" in x:
        return "Penthouse"
    elif "studio" in x or "rk" in x:
        return "Studio"
    else:
        return "Other"

merged_data["property_type"] = merged_data["house_type_clean"].apply(extract_property_type)
merged_data.drop(columns=["house_type"], inplace=True)
```
